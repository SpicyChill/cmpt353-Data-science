## Preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Functions

```python
def get_summary(data: pd.DataFrame, data_name: str):
    r = stats.linregress(data).rvalue
    describe = data.describe()
    summary = pd.DataFrame({
        'x_mean': describe['x']['mean'],
        'y_mean': describe['y']['mean'],
        'x_std': describe['x']['std'],
        'y_std': describe['y']['std'],
        'x_min': describe['x']['min'],
        'y_min': describe['y']['min'],
        'x_max': describe['x']['max'],
        'y_max': describe['y']['max'],
        'r': r,
    }, index=[data_name])
    return summary
```

```python
def data_plot(data: pd.DataFrame):
    plt.plot(data['x'], data['y'], 'b.', alpha=0.5)
    linregress = stats.linregress(data)
    plt.plot(data['x'], data['x'] * linregress.slope + linregress.intercept, 'r-')

```

## Get Summary

```python
i = 1
data = list()
filename = list()
data_prefix = 'data-'
while i <= 6:
    filename.append(data_prefix + str(i) + '.csv')
    data.append(pd.read_csv(filename[i - 1]))
    i += 1
summary = get_summary(data[0], filename[0])
i = 1
while i < len(data):
    summary = summary.append(get_summary(data[i],filename[i]))
    i += 1

```
